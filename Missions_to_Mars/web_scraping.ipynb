```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
from splinter import Browser
```

```python
executable_path = {'executable_path': '/Users/XJY/Desktop/COLNYC20190716DATA-master/01-Lesson-Plans/12-Web-Scraping-and-Document-Databases/2/Activities/08-Stu_Splinter/Unsolved/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = "https://mars.nasa.gov/news/"
browser.visit(url)
```

```python
response = requests.get(url)
soup = bs(response.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
titles = soup.find_all('div', class_="content_title")
print(titles)
```

```python
body = soup.find_all('div', class_="rollover_description")
print(body)
```

```python
results = soup.find_all('div', class_="slide")
for result in results:
    titles = result.find('div', class_="content_title")
    title = titles.find('a').text
    bodies = result.find('div', class_="rollover_description")
    body = bodies.find('div', class_="rollover_description_inner").text
    print('----------------')
    print(title)
    print(body)
```

```python
url = ('https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars')

response = requests.get(url)
soup = bs(response.text, 'html.parser')
```

```python
url = ('https://twitter.com/marswxreport?lang=en')

response = requests.get(url)
soup = bs(response.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
contents = soup.find_all("div",class_="content")
```

```python
weather_mars = []
for content in contents:
    tweet = content.find("div", class_="js-tweet-text-container").text
    weather_mars.append(tweet)
#print(weather_mars)

mars_weather = weather_mars[8]
print(mars_weather)
```

```python
url = ('https://astrogeology.usgs.gov/search/map/Mars/Viking/cerberus_enhanced')

response = requests.get(url)
soup = bs(response.text, 'html.parser')
```

```python
cerberus_img = soup.find_all('div', class_="wide-image-wrapper")
print(cerberus_img)
```

```python
for img in cerberus_img:
    pic = img.find('li')
    full_img = pic.find('a')['href']
    print(full_img)
cerberus_title = soup.find('h2', class_='title').text
print(cerberus_title)
cerberus_hem = {"Title": cerberus_title, "url": full_img}
print(cerberus_hem)
```
